# change the data type of total charges from object to float and replace spaces to 0's.

```python
df['TotalCharges']=df['TotalCharges'].replace(" ",'0')
df['TotalCharges']=df['TotalCharges'].astype('float')
```

```python
df.duplicated("customerID").sum()   # we checked the duplicate based on CustomerID
```

# changed the data type of cenior citizen from (0,1) to (yes or no)

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)   
```

```python
import plotly.express as px

# Count churn values
churn_counts = df['Churn'].value_counts().reset_index()
churn_counts.columns = ['Churn', 'Count']

# Create pie chart
fig = px.pie(
    churn_counts,
    names='Churn',
    values='Count',
    title='Customer Churn Distribution',
    hole=0.3  # Makes it donut style (optional)
)

# Show percentage inside chart
fig.update_traces(textinfo='percent+label')

```

# customer churn by gender

```python
ax=sns.countplot(x='gender', data=df ,hue='Churn')
#ax.bar_label(ax.containers[0])
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Customer Churn Count by Gender")
plt.show()
```

```python
categorical_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

for col in categorical_cols:
    plt.figure(figsize=(8,3))
    sns.countplot(x=col, hue='Churn', data=df)
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.show()
```
